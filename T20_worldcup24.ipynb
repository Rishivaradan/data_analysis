```python
all_teams = pd.concat([matches['team1'], matches['team2']])
total_matches=all_teams.value_counts()
matches_won=matches['winner'].value_counts()
win_percentage=(matches_won/total_matches)*100
print(win_percentage)
```

```python
plt.xlabel("Country")
plt.ylabel("Win percentage")
plt.title("Win percentage of teams")
win_percentage.plot(kind="bar")
```

```python
team_won_batting=matches[(matches['toss_decision'] == 'bat') & (matches['toss_winner'] == matches['winner'])]
team_won_batting.shape[0]
```

```python
team_won_bowling=matches[(matches['toss_decision'] == 'field') & (matches['toss_winner'] == matches['winner'])]
team_won_bowling.shape[0]
```

```python
batting_win_percentage = (team_won_batting.shape[0] / matches.shape[0]) * 100
bowling_win_percentage = (team_won_bowling.shape[0] / matches.shape[0]) * 100
print(f'Percentage of matches won batting first: {batting_win_percentage:.2f}%')
print(f'Percentage of matches won bowling first: {bowling_win_percentage:.2f}%')
```

```python
wicket_takers = deliveries[deliveries['wicket_type'].notnull()]
wicket_takers_count = wicket_takers.groupby('bowler').size().reset_index(name='wickets')
wicket_takers_sorted = wicket_takers_count.sort_values(by='wickets', ascending=False).reset_index(drop=True)
wicket_takers_sorted
```

```python
run_scorers = deliveries.groupby('striker')['runs_off_bat'].sum().sort_values(ascending=False).reset_index(name='total_runs')
run_scorers.head(10)
run_scorers
```
